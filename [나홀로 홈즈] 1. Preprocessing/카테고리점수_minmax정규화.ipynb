# 카테고리 점수 스케일 맞추기   
1. mim-max 정규화를 통해 각 카테고리의 최소-최대값을 0~1로 맞추어준다.
2. x10을 해서 각 카테고리의 최소-최대값을 0~10으로 맞추어준다.

```python
import pandas as pd
```

```python
df = pd.read_excel('data/minmax_연습.xlsx', index_col = 0)
df = df.iloc[0:,16:]
df
```

```python
df.describe()
```

### Min-max 정규화   
(X - MIN) / (MAX-MIN) 라는 식으로 X값의 범위를 0~1로 맞추어 준다.

```python
from sklearn.preprocessing import MinMaxScaler

mMscaler = MinMaxScaler()
normal = mMscaler.fit_transform(df) # transform부터가 0-1으로 바꾸는 것
normal = pd.DataFrame(normal, columns = df.columns)
df = normal*10 #최대값 10으로 맞추기
```

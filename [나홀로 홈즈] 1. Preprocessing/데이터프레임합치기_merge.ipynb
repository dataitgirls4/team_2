# 2. merge로 모든 카테고리의 지표 병합
- how = outer로 조인
- on = dong 코드 일치한다는 조건
- fillna(0) -> N/A값 전부 0으로 전처리

```python
result1 = pd.merge(safety, pet, how='outer', on='dong_code').fillna(0)
result1
```

```python
result2 = pd.merge(result1, sobi, how='outer', on='dong_code').fillna(0)
```

```python
result2
```

```python
result3 = pd.merge(result2, leisure, how='outer', on='dong_code').fillna(0)
result3
```

```python
result4 = pd.merge(result3, selfcare, how='outer', on='dong_code').fillna(0)
result4
```

```python
result5 = pd.merge(result4, traffic, how='outer', on='dong_code').fillna(0)
result5
```

# 1. 좌표로 행정동 뽑기

### step1. 좌표로 지번 주소 뽑기   
- 지번주소 뽑는 이유: 원래는 좌표에서 바로 행정동을 뽑고 싶었음. but 카카오 api에서는 도로명주소와 지번주소만 제공   
- 따라서 좌표로 지번주소 뽑고, 지번주소로 행정동을 뽑는 2step으로 진행


```python
#좌표로 행정주소데이터 불러오는 함수

def getAddress(gy,whi):
    url = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?input_coord=WGS84&output_coord=WGS84&' + 'x=' +gy+ '&y=' +whi
    headers = {"Authorization": "KakaoAK 1eae67e2cf991773e872edc56c0f2b16"}
    api_test = requests.get(url,headers=headers)
    url_text = json.loads(api_test.text)
    return url_text
getAddress('127.150304746486','37.5494403330957')
```

```python
x_y = []
for i in range(len(df)):
    x_y.append((x[i],y[i]))
```

### step2.지번주소로 행정동 찾기

```python
#데이터 프레임에서 주소 리스트 형태로 가져오기

doro = []
for i in range(len(df)) :
    doro.append(df[['지번주소']].values.tolist()[i][0]) #'상세주소'는 불러온 데이터의 칼럼명에 맞게 수정
```

```python
import time
```

```python
#빈리스트 

dong = []

for i in doro:
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={i}'
    headers = {"Authorization": "KakaoAK 1eae67e2cf991773e872edc56c0f2b16"}
    api_test = requests.get(url,headers=headers)
    url_text = json.loads(api_test.text)
    time.sleep(1)
    if not url_text['documents']: #documents가 비어있다면 0으로 채우기 - 도로명 주소로 검색이 제대로 안된 경우
        dong.append(0)
        continue # for문 처음으로 점핑!
    else:
        dong.append(url_text['documents'][0]['address']['region_3depth_h_name'])
```

```python
df['dong']= dong
df.to_excel(excel_writer='피부과일부_파이썬.xlsx')
```
